```python
from src.preprocessing import load_and_clean
from src.split_data import split_data
from src.train_model import train_and_save
from src.evaluate import evaluate
from src.plot_weights import plot_weights
from src.config import DATA_PATH
```

```python
df = load_and_clean(DATA_PATH)
print(f"Data shape: {df.shape}")
print(f"\nFirst few rows:")
df.head()
```

```python
print("Data info:")
print(df.info())
print("\nData statistics:")
df.describe()
```

```python
X_train, X_test, y_train, y_test = split_data(df)
print(f"Training set: X={X_train.shape}, y={y_train.shape}")
print(f"Test set: X={X_test.shape}, y={y_test.shape}")
print(f"\nFeature names: {list(X_train.columns)}")
```

```python
model, weights, model_path = train_and_save(X_train, y_train)
print(f"Model saved at: {model_path}")
print(f"\nModel coefficients (weights):")
for i, (feature, weight) in enumerate(zip(X_train.columns, weights)):
    print(f"  {feature}: {weight:.6f}")
```

```python
results = evaluate(model, X_test, y_test)
print("Evaluation results:")
for metric, value in results.items():
    print(f"  {metric.upper()}: {value:.6f}")
```
